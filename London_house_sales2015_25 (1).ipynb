````python
#import pandas to analyse this data
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
house_sales = pd.read_csv('/content/drive/MyDrive/ppd_data.csv')
house_sales

````

````python

hsd = house_sales.copy()
hsd = hsd.drop(columns =['linked_data_uri'])
hsd
````

````python
print(hsd['district'].unique())
print(len(hsd['district'].unique()))
````

Checking which burrough had the most number of sales.

````python
boroughs = hsd['district'].value_counts()
plt.figure(figsize=(12,6))
max_value = boroughs.values.max()
plt.bar(boroughs.index,boroughs.values)
plt.xticks(rotation=90)
plt.title("Number Of Sales In Each Borough")
plt.yticks(np.arange(0,max_value+1000,1000))
plt.show()
````

````python
avg_price=(hsd.groupby('district')['price_paid'].mean().reset_index())
avg_price=avg_price.sort_values('price_paid')
#print (avg_price)
plt.figure(figsize=(12,6))
plt.plot(avg_price['district'],avg_price['price_paid'],marker='o')
plt.xticks(rotation=90);
plt.yticks(np.arange(0,avg_price['price_paid'].max(),100000));
plt.title('Average Price of a Property Across The Burroughs In London');
plt.ylabel('Million GBP');
````

````python
hsd['deed_date'] = pd.to_datetime(hsd['deed_date'])
hsd['Year']= hsd['deed_date'].dt.year
hsd['Month'] = hsd['deed_date'].dt.month_name()
hsd


````

````python
annual_price=hsd.groupby('Year')['price_paid'].mean().reset_index()
plt.figure(figsize=(12,6))
plt.plot(annual_price['Year'],annual_price['price_paid'], marker='o')
plt.xticks(annual_price['Year']);
plt.grid(True)
plt.yticks(np.arange(600000,annual_price['price_paid'].max()+100000,50000));
plt.show()

#plt.yticks(np.arange(0,annual_price['price_paid'].max(),100000));
````

Do the flat sales correlate with the total property sold in that burrough?

````python
number = hsd['property_type'].value_counts()

plt.pie(number,labels = number.index,autopct='%1.1f%%',
    startangle=90);
````

Plot property type sales in each burrough

````python
pt = hsd.pivot_table(index='district',
              columns = 'property_type',
              values='price_paid',
              aggfunc='count',
              fill_value=0)
names={'D':'Detached',
       'F':'Flats',
       'S':'Semi-Detached',
       'T':'Terraced'}
pt = pt.rename(columns=names)
pt['total_sales']=pt.sum(axis=1)
pt= pt.sort_values('total_sales',ascending = False)
pt =pt.drop(columns= 'total_sales')
pt
````

````python
plt.figure(figsize=(12,6))
bottom=np.zeros(len(pt.index))

for col in pt.columns:
  plt.bar(pt.index,pt[col],bottom = bottom, label = col)
  bottom+=pt[col].values

plt.xticks(rotation=90)
plt.legend()
plt.yticks(np.arange(0,bottom.max()+1000,1000));
plt.title('SALES IN EACH BOROUGH BY PROPERTY TYPE')
````
